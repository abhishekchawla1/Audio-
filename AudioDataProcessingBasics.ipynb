```python
path='/content/drive/MyDrive/Human Audio/Actor_04/03-01-01-01-02-02-04.wav'
```

```python
import matplotlib.pyplot as plt
from IPython.display import Audio
import librosa
import numpy as np
```

```python
Audio(path)
```

```python
data,sr=librosa.load(path)
```

```python
import librosa.display
```

```python
plt.figure(figsize=(14, 5))
plt.plot(data)
plt.title('Waveplot')
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.show()
```

```python
data.shape
```

```python
#Mel Frequency Cepstral Coefficient

#Freq and Time components
```

```python
mfcc=librosa.feature.mfcc(y=data,sr=sr,n_mfcc=40)
```

```python
mfcc
```

```python
mfcc.shape
```

```python
mfcc_scaled_features=np.mean(mfcc.T,axis=0)
```
